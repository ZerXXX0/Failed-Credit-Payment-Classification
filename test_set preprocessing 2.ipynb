```python
import pandas as pd

# Load train for encoder fitting and test for transformation
train_df_enc = pd.read_csv('adikara-2025-indonesia-kredit-macet/train.csv')
test_df = pd.read_csv('adikara-2025-indonesia-kredit-macet/test.csv')

# Frequency encoder helper: returns normalized frequency map
def build_freq_map(series):
    counts = series.value_counts(dropna=False)
    total = counts.sum()
    return (counts / total).to_dict()

# Columns to frequency-encode (categoricals)
cat_cols = ['provinsi', 'jenis_pinjaman', 'status_peminjam', 'sektor_usaha', 'pendidikan', 'jenis_jaminan']

# Build frequency maps from train and apply to test
freq_maps = {}
for col in cat_cols:
    if col in train_df_enc.columns and col in test_df.columns:
        freq_maps[col] = build_freq_map(train_df_enc[col])
        # Simple if-else style mapping with fallback to 0 for unseen/missing
        test_df[col] = test_df[col].map(lambda x: freq_maps[col][x] if x in freq_maps[col] else 0)

# Impute missing numeric values with 0 (keep prior behavior)
for num_col in ['total_pengembalian', 'porsi_pengembalian_lender', 'durasi_hari']:
    if num_col in test_df.columns:
        test_df[num_col] = test_df[num_col].fillna(0)

# Convert tanggal_pencairan to days since 2000-01-01 if present
if 'tanggal_pencairan' in test_df.columns:
    base_date = pd.Timestamp('2000-01-01')
    test_df['tanggal_pencairan'] = pd.to_datetime(test_df['tanggal_pencairan'], errors='coerce')
    test_df['tanggal_pencairan'] = (test_df['tanggal_pencairan'] - base_date).dt.days.fillna(0)

# Save preprocessed test
test_df.to_csv('test_preprocessed with impute 0.csv', index=False)
```

```python
import numpy as np
import pandas as pd
from category_encoders import CatBoostEncoder

# Columns to encode
cat_cols = ['provinsi', 'jenis_pinjaman', 'status_peminjam', 'sektor_usaha', 'pendidikan', 'jenis_jaminan']

# Load raw train and test (for consistent, leakage-safe encoders)
train = pd.read_csv('adikara-2025-indonesia-kredit-macet/train.csv')
test = pd.read_csv('adikara-2025-indonesia-kredit-macet/test.csv')

# Minimal alignment with train preprocessing
# Drop rows in train like earlier cleaning steps (do NOT drop rows in test)
required_non_null = ['total_pengembalian', 'porsi_pengembalian_lender', 'jenis_pinjaman', 'durasi_hari']
train = train.dropna(subset=required_non_null)

# Convert tanggal_pencairan to numeric days since 2000 for both train and test
base_date = pd.Timestamp('2000-01-01')
for df in (train, test):
    if 'tanggal_pencairan' in df.columns:
        df['tanggal_pencairan'] = pd.to_datetime(df['tanggal_pencairan'], errors='coerce')
        df['tanggal_pencairan'] = (df['tanggal_pencairan'] - base_date).dt.days

# 1) Frequency encoding: compute from TRAIN only, map to TEST
for c in cat_cols:
    if c in train.columns and c in test.columns:
        freqs = train[c].value_counts(normalize=True)
        test[c + '_freq'] = test[c].map(freqs).astype(float).fillna(0.0)

# 2) Target encoding for TEST using CatBoostEncoder fitted on TRAIN
#    (OOF is for train; for test we fit on full train and transform test)
enc = CatBoostEncoder(cols=cat_cols)
enc.fit(train[cat_cols], train['status_gagal_bayar'])
te_test = enc.transform(test[cat_cols])
for c in cat_cols:
    test[c + '_te'] = te_test[c]

# Finalize test set: drop raw categorical columns, keep id_transaksi for submission join
drop_raw = [c for c in cat_cols if c in test.columns]
output_test = test.drop(columns=drop_raw, errors='ignore')

# Save
output_path = 'test_preprocessed_with_freq_te_and_columns_dropped.csv'
output_test.to_csv(output_path, index=False)
print('Saved:', output_path)
output_test.head()
```
