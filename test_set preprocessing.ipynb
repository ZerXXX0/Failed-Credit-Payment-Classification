```python
import pandas as pd

# Load train for encoder fitting and test for transformation
train_df_enc = pd.read_csv('/Users/ailab/Documents/gojiii/adikara-2025-indonesia-kredit-macet/train.csv')
test_df = pd.read_csv('/Users/ailab/Documents/gojiii/adikara-2025-indonesia-kredit-macet/test.csv')

# Frequency encoder helper: returns normalized frequency map
def build_freq_map(series):
    counts = series.value_counts(dropna=False)
    total = counts.sum()
    return (counts / total).to_dict()

# Columns to frequency-encode (categoricals)
cat_cols = ['provinsi', 'jenis_pinjaman', 'status_peminjam', 'sektor_usaha', 'pendidikan', 'jenis_jaminan']

# Build frequency maps from train and apply to test
freq_maps = {}
for col in cat_cols:
    if col in train_df_enc.columns and col in test_df.columns:
        freq_maps[col] = build_freq_map(train_df_enc[col])
        # Simple if-else style mapping with fallback to 0 for unseen/missing
        test_df[col] = test_df[col].map(lambda x: freq_maps[col][x] if x in freq_maps[col] else 0)

# Impute missing numeric values with 0 (keep prior behavior)
for num_col in ['total_pengembalian', 'porsi_pengembalian_lender', 'durasi_hari']:
    if num_col in test_df.columns:
        test_df[num_col] = test_df[num_col].fillna(0)

# Convert tanggal_pencairan to days since 2000-01-01 if present
if 'tanggal_pencairan' in test_df.columns:
    base_date = pd.Timestamp('2000-01-01')
    test_df['tanggal_pencairan'] = pd.to_datetime(test_df['tanggal_pencairan'], errors='coerce')
    test_df['tanggal_pencairan'] = (test_df['tanggal_pencairan'] - base_date).dt.days.fillna(0)

# Save preprocessed test
test_df.to_csv('test_preprocessed with impute 0.csv', index=False)
```
