```python
import pandas as pd

# Load datasets
train_raw = pd.read_csv("adikara-2025-indonesia-kredit-macet/train.csv")
train_enc = pd.read_csv("train_preprocessed.csv")
test = pd.read_csv("adikara-2025-indonesia-kredit-macet/test.csv")

# List of categorical columns to map
categorical_cols = [
    'provinsi', 'jenis_pinjaman', 'status_peminjam',
    'sektor_usaha', 'pendidikan', 'jenis_jaminan'
]

# Build raw→encoded mapping from train.csv and train_preprocessed.csv
mapping = {}
for col in categorical_cols:
    if col in train_raw.columns and col in train_enc.columns:
        mapping[col] = dict(zip(train_raw[col], train_enc[col]))

# Apply simple if–else encoding to test.csv
for col in categorical_cols:
    if col in test.columns:
        test[col] = test[col].apply(lambda x: mapping[col][x] if x in mapping[col] else -1)

test.head()
```

```python
# Convert to datetime
test['tanggal_pencairan'] = pd.to_datetime(test['tanggal_pencairan'])

# Base date
base_date = pd.Timestamp('2000-01-01')

# Calculate days after year 2000
test['tanggal_pencairan'] = (test['tanggal_pencairan'] - base_date).dt.days
```
